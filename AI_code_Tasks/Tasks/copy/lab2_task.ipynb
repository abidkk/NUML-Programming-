### 3.6.2 LAB TASK 2:
- Create a class name basic_calc with following attributes and methods;
- Two integers (values are passed with instance creation)
- Different methods such as addition, subtraction, division, multiplication

```python
class BasicCalc:
    # constructor
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    # Methods 1:  SUM
    def sum(self):
        print(f"SUM of Two Numers: {self.num1+self.num2}")

    # Methods 2:  Subtract
    def subtract(self):
        print(f"Subtraction of Two Numers: {self.num1 - self.num2}")

    # Method 3:  Multiplication
    def multi(self):
        print(f"Multiplicaiton of Two Numers: {self.num1 * self.num2}")

       # Method 4:  Division
    def division(self):
        print(f"Division of Two Numers: {self.num1 / self.num2}")
    
# object or instances
obj1 = BasicCalc(200,10)
obj1.sum()
obj1.subtract()
obj1.multi()
obj1.division()
```
