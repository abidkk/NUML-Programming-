##### Task 4: Model-Based Warehouse Robot Agent

```python
class WarehouseAgent:

    def __init__(self):
        self.picked_items = {"p1":None, "p2":None}
        self.current_location = None

    def decide(self, percept):
        location, has_package = percept

        #  updating internal 
        self.current_location = location
        #package exists 
        if has_package and not self.picked_items[location]:
            self.picked_items[location] = True
            return "Pick"

        #  already picked 
        if self.picked_items[location]:
            return "Skip"

        return "Move"
# instance 
agent1 = WarehouseAgent()
```

```python
print(agent1.decide(("p1", False)))
```

```python
print(agent1.decide(("p2", True)))
print(agent1.decide(("p1", True)))
print(agent1.decide(("p2", False)))
```
