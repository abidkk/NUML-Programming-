##### Task 1: Simple Reflex Vacuum Cleaner Agent 

```python
class VacumAgent:

    def decide(self, percept):
        location, status = percept

        if (status.lower() == "dirty"):
            return "Clean"
        elif (status.lower() == "clean"):
            return "Move"
        else:
            return "Invalid Input"

agent1 = VacumAgent()
```

```python
print(agent1.decide(("A", "dirty")))
print(agent1.decide(("A", "clean")))
```
