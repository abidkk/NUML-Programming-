##### Task 3: Model-Based Vacuum Cleaner Agent

```python
class ModelVacuum:
    
    def __init__(self):
        self.state = {"A": "None", "B": "None"}
        self.current_location = None

    def decide(self, percept):

        location, status = percept

        self.current_location = location

        # updating state
        self.state[location] = status

        #check both rooms
        if self.state["A"] == "Clean" and self.state["B"] == "Clean":
            return "NoAction"
        
        if status == "Dirty":
            return "Suck"
        elif status == "Clean":
            return "Move"

# object
agent1 = ModelVacuum()
```

```python
print(agent1.decide(("A", "Dirty"))) # Suck
print(agent1.decide(("A", "Clean"))) # Move
print(agent1.decide(("B", "Dirty"))) # Suck
print(agent1.decide(("B", "Clean"))) # N Action
```
