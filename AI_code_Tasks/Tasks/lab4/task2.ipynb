##### Task 2: Traffic Light Reflex Agent 

```python
class TrafficAgent:
    
    def decide(self, light_color):

        if (light_color.lower() == "red"):
            return "STOP"
        elif (light_color.lower() == "yellow"):
            return "SLOW"
        elif (light_color.lower() == "green"):
            return "GO"
        else:
            return "Invalid Input"
        
agent1 = TrafficAgent()
```

```python
print(agent1.decide("green")) # go
print(agent1.decide("rED")) # Stop
print(agent1.decide("YELLOW")) # Slow
print(agent1.decide("orange")) # invalid Input
```
