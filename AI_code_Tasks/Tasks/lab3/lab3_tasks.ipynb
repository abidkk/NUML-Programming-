# Lab Task 1:
Write a program to Calculate the length of a String without using built-int len() Function

```python
def calcStringLen(msg):
    try:
        count = 0
        for _ in msg:
            count +=1
        return f"The Length of '{msg}' is: {count}"
    except Exception as e:
        print(e)

print(calcStringLen("Abid")) # 4 
print(calcStringLen("BSCS KC 006")) # 11

print(calcStringLen(2)) # handling error on non string value
```

# Lab Task 2 
Write a program that creates a list of 10 random integers. 
Then create Two lists by name odd_list and even_list that holds odd and even values respectively

```python
import random

def num_lists():
    odd_list= []
    even_list = []
    for _ in range(10):
        nums = random.randint(20, 60) # generate 10 random nums bw 20 to 60
        if nums %2 == 0:
            even_list.append(nums)
        else:
            odd_list.append(nums)

    # return [f"Even List: {even_list}", f"Odd List: {odd_list}"] # returns List
    print(f"Even List: {even_list}")
    print(f"Odd List: {odd_list}")

num_lists()
```
